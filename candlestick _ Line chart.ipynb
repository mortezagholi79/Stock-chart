# candlestick _ Line chart

```python
# Importing packages

import pandas as pd
import datetime as dt
import pandas_datareader.data as web
import plotly.express as px
import plotly.graph_objects as go
```

```python


def stock_chart(chart , Year , Month , Day , Type):
    start = dt.datetime(Year,Month,Day)
    end = dt.datetime.now()
    stocks = web.DataReader([chart], 'yahoo', start, end)
    if 'candlestick' in Type:
        candlestick = go.Figure(data = [go.Candlestick(x = stocks.index, open = stocks[('Open',    chart)], high = stocks[('High',    chart)], 
                                               low = stocks[('Low',    chart)], close =stocks[('Close',   chart)] )])
        candlestick.update_layout(xaxis_rangeslider_visible = False)
        candlestick.update_xaxes(title_text = 'Date')
        candlestick.update_yaxes(title_text = ' Close Price', tickprefix = '$')
        candlestick.show()
    if 'Line chart' in Type :
        stocks_close = pd.DataFrame(web.DataReader([chart], 'yahoo', start, end)['Close'])
        area_chart = px.area(stocks_close)
        area_chart.update_xaxes(title_text = 'Date')
        area_chart.update_yaxes(title_text = ' Close Price', tickprefix = '$')
        area_chart.update_layout(showlegend = False)
        area_chart.show()
```

```python
stock_chart('AAPL',2019,2,2 , 'candlestick')
```

```python
stock_chart('AAPL',2019,2,2 , 'Line chart')
```
